```python
import pandas as pd
from fuzzywuzzy import fuzz, process
from haversine import haversine

# Load the dataset
df = pd.read_csv('otter_data_cleaning.csv', sep='\u0001')
df = df[~df.restaurant_id.isna()]
df = df.dropna(subset=['name'])
df['latitude'] = df['latitude'].astype('float')
df['longitude'] = df['longitude'].astype('float')
df['restaurant_id'] = df['restaurant_id'].astype('object') 


# 1. Data Preparation:

def standardize_strings(column):
    return column.str.lower().str.strip().str.replace('[^\w\s]', '')

def round_geolocation(df):
    df['latitude'] = df['latitude'].round(4)
    df['longitude'] = df['longitude'].round(4)
    return df

df['name'] = standardize_strings(df['name'])
df['standardized_name'] = standardize_strings(df['standardized_name'])
df = round_geolocation(df)

# 2. Feature Engineering:

def extract_core_name(name):
    # Additional or different keywords can be added based on the dataset specifics
    keywords = ['downtown', 'west side', 'east side', 'north', 'south']
    for word in keywords:
        name = name.replace(word, '')
    return name.strip()

df['core_name'] = df['name'].apply(extract_core_name)

def calculate_similarity(row, column_name, choices):
    best_match = process.extractOne(row[column_name], choices)
    if best_match:
        return best_match[1]
    return 0

all_names = df['core_name'].unique().tolist()
df['similarity_score'] = df.apply(calculate_similarity, args=('core_name', all_names), axis=1)

df['name_tokens'] = df['name'].str.split()

def calculate_avg_distance(row):
    location = (row['latitude'], row['longitude'])
    all_other_locations = df[(df['restaurant_id'] != row['restaurant_id'])][['latitude', 'longitude']].values.tolist()
    if not all_other_locations:
        return 0
    avg_distance = sum(haversine(location, (lat, lon)) for lat, lon in all_other_locations) / len(all_other_locations)
    return avg_distance

df['avg_distance_to_others'] = df.apply(calculate_avg_distance, axis=1)

# Save the modified dataframe
# df.to_csv('processed_restaurant_data.csv', index=False)
```
